```python
from tensorflow import keras
from keras.applications.vgg16 import VGG16,decode_predictions
```

```python
model = VGG16()
```

```python
model.summary()
```

#### Dog Prediction

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/dog.jpg")
```

```python
plt.imshow(img)
```

```python
img = cv2.resize(img,(224,224))
```

```python
img_input = img.reshape(1,224,224,3)
```

```python
y_pred = model.predict(img_input)
decode_predictions(y_pred,top=1)
```
