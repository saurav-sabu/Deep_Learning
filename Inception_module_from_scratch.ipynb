```python
from keras.layers import *
from keras.utils import plot_model
from keras.models import Model
```

```python
input = Input(shape=(32,32,3))
```

```python
tower_1 = Conv2D(64,kernel_size=(1,1),padding="same",activation="relu")(input)
tower_1 = Conv2D(64,kernel_size=(3,3),padding="same",activation="relu")(tower_1)

tower_2 = Conv2D(64,kernel_size=(1,1),padding="same",activation="relu")(input)
tower_2 = Conv2D(64,kernel_size=(5,5),padding="same",activation="relu")(tower_2)

tower_3 = MaxPooling2D(pool_size=(3,3),padding="same",strides=(1,1))(input)
tower_3 = Conv2D(64,(1,1),padding="same",activation="relu")(tower_3)
```

```python
output = concatenate([tower_1,tower_2,tower_3],axis=3)
```

```python
flat = Flatten()(output)
final = Dense(10,activation="softmax")(flat)
```

```python
model = Model(inputs=input,outputs=final)
```

```python
plot_model(model)
```
